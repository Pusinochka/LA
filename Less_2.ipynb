**2.2. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –º–∞—Ç—Ä–∏—Ü—ã $A$ —Ä–∞–≤–µ–Ω $4$. –ù–∞–π–¥–∏—Ç–µ:**

   –∞) $det(A^{2})$;
    
   –±) $det(A^{T})$;
    
   –≤) $det(2A)$.

a) –°–æ–≥–ª–∞—Å–Ω–æ —Å–≤–æ–π—Å—Ç–≤—É –¥–ª—è –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü –æ–¥–∏–Ω–∞–∫–æ–≤–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞,
$det(AB)=detA\cdot detB.$<br> 
–ü–æ—ç—Ç–æ–º—É $det(A^{2})$ = $det(A)*det(A)$ = 16.

–±) $det(A^{T})$;<br>
–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã —Ä–∞–≤–µ–Ω –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—é –∏—Å—Ö–æ–¥–Ω–æ–π: $detA^{T}=detA.$ = 16,

–≤) $det(2A)$
–£–º–Ω–æ–∂–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏ –∏–ª–∏ —Å—Ç–æ–ª–±—Ü–∞ –º–∞—Ç—Ä–∏—Ü—ã –Ω–∞ —á–∏—Å–ª–æ $\lambda$ –ø—Ä–∏–≤–µ–¥–µ—Ç –∫ —É–º–Ω–æ–∂–µ–Ω–∏—é –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –º–∞—Ç—Ä–∏—Ü—ã –Ω–∞ —Ç–æ –∂–µ —á–∏—Å–ª–æ. –ó–Ω–∞—á–∏—Ç, –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã –ø–æ—Ä—è–¥–∫–∞ n, n —Å—Ç—Ä–æ–∫ –±—É–¥—É—Ç —É–º–Ω–æ–∂–µ–Ω—ã –Ω–∞ 2, –∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å —É–≤–µ–ª–∏—Ç—Å—è –≤ 2^{n} —Ä–∞–∑.<br>
$det(2A) = detA *2^{n} = 16$ 

```python
a = np.array([[5, 2], [3, 2]]) #–≤–æ–∑—å–º—ë–º –¥–ª—è –ø—Ä–∏–º–µ—Ä–∞ —Ç–∞–∫—É—é –º–∞—Ç—Ä–∏—Ü—É
b = np.dot(a, a)
c = 2*a
print(f'–ú–∞—Ç—Ä–∏—Ü–∞:\n{b}')
print(f'–ú–∞—Ç—Ä–∏—Ü–∞:\n{c}')
print(f'–∞) {np.linalg.det(b):.0f}')
print(f'–±) {np.linalg.det(a.T):.0f}')
print(f'–≤) {np.linalg.det(c):.0f}')
```

**2.4. –ù–∞–π–¥–∏—Ç–µ —Ä–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã:**

   –∞) $\begin{pmatrix}
1 & 2 & 3\\ 
1 & 1 & 1\\ 
2 & 3 & 4
\end{pmatrix};$

   –±) $\begin{pmatrix}
0 & 0 & 2 & 1\\ 
0 & 0 & 2 & 2\\ 
0 & 0 & 4 & 3\\

\end{pmatrix}.$

```python
x = [1, 2, 3]
y = [1, 1, 1]
z = [2, 3, 4]

a = np.array([x, y, z])
r = np.linalg.matrix_rank(a)

print(f'–†–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã a: {r}')
```
