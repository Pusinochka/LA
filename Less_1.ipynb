$$\quad x = (0.5\cdot b1,1\cdot b2,3\cdot b3)=
(\frac{1}{2}\cdot b_1, b_2, 3\cdot b_3) = (0.5, 1, 3)$$

```python
b1 = np.array([0, 0, 10])
b2 = np.array([2, 0, 0])
b3 = np.array([0, 1, 0])
x = np.array([2, 3, 5])
```

```python
z = np.dot(y_1, x)
z
```

**2.2._ –ù–∞–π–¥–∏—Ç–µ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ $(4, 2, 4)$ –∏ $(12, 3, 4)$ –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.**

```python
x = np.array([4, 2, 4])
y = np.array([12, 3, 4])
print(f'–ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ x:  {norm(x, ord=1)}')
print(f'–ú–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–∞—è –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ y:  {norm(y, ord=1)}')
print(f'–ï–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ x:  {norm(x, ord=2)}')
print(f'–ï–≤–∫–ª–∏–¥–æ–≤–∞ –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ y:  {norm(y, ord=2)}')
```

```python
cos_phi = np.dot(x, y) / norm(x) / norm(y)
print(f'–ö–æ—Å–∏–Ω—É—Å —É–≥–ª–∞ –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏:  {cos_phi}')
print(f'–£–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏: {np.arccos(cos_phi)}')
```

**2.4. –í—ã—è—Å–Ω–∏—Ç–µ, –∫–∞–∫–∏–µ –∏–∑ –Ω–∏–∂–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –ª–∏–Ω–µ–π–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ $\mathbb{R}^{3}$**:<br>
–∞) $(1,0,0),(0,0,1);$ <br>

–î–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞ –Ω–µ –º–æ–≥—É—Ç –æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –ª–∏–Ω–µ–π–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ $\mathbb{R}^{3}$

–±) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$

```python
x = np.array([1/np.sqrt(2), -1/np.sqrt(2), 0])
y =  np.array([1/np.sqrt(2), 1/np.sqrt(2), 0])
z = np.array([0, 0, 1])

print(f'—Ç.–∫. —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ {np.dot(x, np.dot(y, z))}, –≤–µ–∫—Ç–æ—Ä–∞ x, y –∏ z –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å')
```

–≤) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$

```python
x_2 = np.array([1/2, -1/2, 0])
y_2 =  np.array([0, 1/2, 1/2])
z_2 = np.array([0, 0, 1])
evc_2 = x_2[0] * y_2[0] * z_2[0] + x_2[1] * y_2[1] * z_2[1] + x_2[2] * y_2[2] * z_2[2]

print(f'—Ç.–∫. —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ {np.dot(x_2, np.dot(y_2, z_2))}, –≤–µ–∫—Ç–æ—Ä–∞ x_2, y_2 –∏ z_2 –Ω–µ –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å')
```
